# Regress√£o Linear em Python

## Introdu√ß√£o

A **Regress√£o Linear** √© uma das t√©cnicas estat√≠sticas e de machine learning mais simples e mais utilizadas para problemas de **previs√£o num√©rica**.

> Ela busca encontrar uma **rela√ß√£o linear** entre uma vari√°vel dependente (o que queremos prever) e uma ou mais vari√°veis independentes (os fatores que influenciam).

> Essa formula√ß√£o tamb√©m √© base para outros modelos de regress√£o em diversos contextos, como Modelos Lineares Genaralizados (MLG) e regress√£o em modelos de s√©ries temporais como o SARIMA-X.
---

### Exemplos onde a regress√£o linear √© usada:

- Previs√£o de pre√ßos de im√≥veis com base em caracter√≠sticas como tamanho, localiza√ß√£o e n√∫mero de quartos.
- Estimativa de vendas em fun√ß√£o do investimento em marketing.
- Previs√£o de consumo el√©trico em fun√ß√£o da temperatura.
---


---
## Formula√ß√£o Matem√°tica B√°sica

### Regress√£o Linear Simples

A **Regress√£o Linear Simples** envolve apenas uma vari√°vel independente (exemplo: √°rea de um im√≥vel).

A equa√ß√£o da reta pode ser representada como:

$$
y_i = \beta_0 + \beta_1 \cdot x_i + …õ_i,
$$

em que, $\beta_0 \in \mathbb{R}$ e $\beta_1 \in \mathbb{R}$, sendo:

- $y$: vari√°vel dependente (target, vari√°vel resposta, o que queremos prever);
- $x$: vari√°vel independente (feature ou covari√°vel);
- $\beta_0$: intercepto (valor de $y$ quando $x = 0$);
- $\beta_1$: coeficiente angular (quanto $y$ varia a cada unidade de $x$);
- $…õ$: termo de erro;

---

### Regress√£o Linear M√∫ltipla

Quando temos **mais de uma vari√°vel independente**, usamos a **Regress√£o Linear M√∫ltipla**.

A equa√ß√£o geral √©:

$$
y_i = \beta_0 + \beta_1 \cdot x_{1,i} + \beta_2 \cdot x_{2,i} + \dots + \beta_p \cdot x_{p,i} + …õ_i
$$

em que, $( \beta_0, \beta_1, \beta_2, \dots, \beta_p )' \in \mathbb{R}^{p+1}$, sendo:

- $( 1, x_1, x_2, \dots, x_p )'$: vetor de vari√°veis independentes (o valor 1 √© adicionado por conta do intercepto);
- $( \beta_0, \beta_1, \beta_2, \dots, \beta_p  )'$: vetor de coeficientes associados a cada vari√°vel.


---

### Principais Suposi√ß√µes da Regress√£o Linear

Antes de aplicar a regress√£o linear, algumas suposi√ß√µes devem ser consideradas:

- **Linearidade**: A rela√ß√£o entre as vari√°veis independentes e dependente √© linear.
- **Independ√™ncia dos erros**: Os erros s√£o independentes.
- **Homoscedasticidade**: Vari√¢ncia constante dos erros.
- **Normalidade dos erros**: Os erros seguem uma distribui√ß√£o Normal.

> As ultimas tr√™s podem ser resumidas em $\varepsilon_1, \dots, \varepsilon_n$  i.i.d. Normal($0, \sigma^2$).
---

## Vantagens e Limita√ß√µes

### Vantagens:
- Simplicidade e facilidade de interpreta√ß√£o.
- Baixo custo computacional.
- Serve como uma boa baseline para problemas de regress√£o.

### Limita√ß√µes:
- N√£o captura rela√ß√µes n√£o lineares de forma eficiente.
- Sens√≠vel a outliers.
- Depende da validade das suposi√ß√µes estat√≠sticas.

---



---
## Formula√ß√£o Matricial e Estimadores

Para problemas com m√∫ltiplas vari√°veis independentes, √© comum representar a regress√£o linear de forma **matricial**.

---

### Representa√ß√£o Matricial

A equa√ß√£o geral da regress√£o linear m√∫ltipla pode ser reescrita como:
$$
\mathbf{y} = \mathbf{X} \boldsymbol{\beta} + \boldsymbol{\epsilon}
$$
em que:

- $\mathbf{y}$ = vetor coluna de tamanho $n \times 1$ com as observa√ß√µes da vari√°vel dependente.
- $\mathbf{X}$ = matriz de tamanho $n \times (p+1)$, onde:
  - Cada linha representa uma observa√ß√£o.
  - Cada coluna representa uma vari√°vel independente (incluindo uma coluna de 1s para o intercepto).
- $\boldsymbol{\beta}$ = vetor dos coeficientes do modelo $(p+1) \times 1$.
- $\boldsymbol{\epsilon}$ = vetor de erros (res√≠duos).

---

### Estimador dos M√≠nimos Quadrados Ordin√°rios (OLS)

O objetivo √© encontrar o vetor de coeficientes $\boldsymbol{\beta}$ que **minimiza a soma dos quadrados dos res√≠duos**, isto √©,
$$
\hat{\boldsymbol{\beta}} = \arg \min \,\sum_{i=1}^n \varepsilon_i^2 = \arg \min \,(\mathbf{y} - \mathbf{X} \boldsymbol{\beta})^\top \,(\mathbf{y} - \mathbf{X} \boldsymbol{\beta})
$$

A solu√ß√£o anal√≠tica do problema de m√≠nimos quadrados √© dada por:
$$
\hat{\boldsymbol{\beta}} = (\mathbf{X}^\top \mathbf{X})^{-1} \mathbf{X}^\top \mathbf{y}
$$
em que:
- $\hat{\boldsymbol{\beta}}$ = vetor dos coeficientes estimados;
- $\mathbf{X}^\top$ = transposta da matriz $\mathbf{X}$;
- $(\mathbf{X}^\top \mathbf{X})^{-1}$ = inversa da matriz $\mathbf{X}^\top \mathbf{X}$.

<br><br>
> Sob a suposi√ß√£o de normalidade do termo de erro $(\varepsilon_i \sim^{i.i.d.} Normal(0,\sigma^2))$, pode-se mostrar que o Estimador M√°xima Verossimilhan√ßa (EMV) coincide com o OLS.

<br><br>
### Lista de estimadores alternativos:

| Tipo de Estimador                                           | Caracter√≠sticas                                                                      |
| ----------------------------------------------------------- | ------------------------------------------------------------------------------------ |
| **OLS (Ordinary Least Squares)**                            | Assume erros independentes, homoced√°sticos, sem penaliza√ß√£o                          |
| **WLS (Weighted Least Squares)**                            | Atribui pesos diferentes para cada observa√ß√£o (quando vari√¢ncias n√£o s√£o constantes) |
| **GLS (Generalized Least Squares)**                         | Lida com erros que podem ser correlacionados ou heteroced√°sticos                     |
| **Ridge Regression (M√≠nimos Quadrados com Penaliza√ß√£o L2)** | Inclui penalidade para reduzir overfitting                                           |
| **Lasso Regression (M√≠nimos Quadrados com Penaliza√ß√£o L1)** | Penaliza o valor absoluto dos coeficientes (sparse solutions)                        |

## Pr√©-processamento

Nesta etapa vamos fazer um tratamento no conjunto de dados j√° pensando na modelagem e na avalia√ß√£o dos modelos.

### Separando target e features

````python
# Target, vari√°vel dependente ou vari√°vel resposta
y = df['sales']

# Features, vari√°veis independentes ou covari√°veis
X = df.drop('sales',axis=1)
````

### Separando treinamento e teste

Em modelo de regress√£o √© uma boa pr√°tica separar uma parte do conjunto de dados como teste, ou seja, vamos separar um subconjunto dos dados que ser√° utilizado para avaliar o modelo fora da amostra de ajuste.

> Tamb√©m existem t√©cnicas mais gerais e robustas, especialmente utilizadas para conjuntos de dados menores, como *K-fold cross-validation* e *Leave-One-Out Cross-Validation*.

No exemplo, vamos separar:
- 70% para o treinamento (ajuste do modelo);
- 30% para o teste (avalia√ß√£o do modelo fora da amostra).

````python
from sklearn.model_selection import train_test_split

# Dividir os dados
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Tamanho do conjunto de treino: {X_train.shape}")
print(f"Tamanho do conjunto de teste: {X_test.shape}")
````

### Padronizando features

Diversos modelos de regress√£o apresentam melhores resultados quando as vari√°veis possuem a escala padronizada, ou seja, √© aplicada uma transforma√ß√£o do tipo $z = (x - \bar{x})/s$ para cada covari√°vel.

- Isto faz com que cada covari√°vel tenha m√©dia zero e vari√¢ncia 1;

- Ap√≥s essa transforma√ß√£o, a varia√ß√£o de uma unidade em uma covari√°vel corresponder√° a varia√ß√£o de um desvio-padr√£o na escala original.

````python
from sklearn.preprocessing import StandardScaler

# Instanciando o scaler
scaler = StandardScaler()

# Ajustando o scaler com os dados de treino e transformando ambos os conjuntos
# -> retorna numpy.ndarray
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Retorna o formato para dataframe
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns, index=X_train.index)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns, index=X_test.index)
````

## Regress√£o Linear Simples

Neste exemplo, vamos relacionar a vari√°vel resposta (`sales`) com a vari√°vel explicativa `TV`, ou seja, vamos utilizar um modelo com a seguinte equa√ß√£o:

$$
sales_i = \beta_0 + \beta_1 \times TV_i + …õ_i,
$$

em que, $\beta_0 \in \mathbb{R}$, $\beta_1 \in \mathbb{R}$ e $…õ_1, \dots, …õ_n$ s√£o i.i.d. Normal$(0,\sigma^2)$.


> OBS: lembre que n√≥s padronizados a vari√°vel `TV`, ou seja, o modelo proposto vai medir o quanto a varia√ß√£o de um desvio padr√£o em `TV` afeta `sales`.

### Gr√°fico de dispers√£o

````python
plt.figure(figsize=(6, 3))
plt.scatter( x=X_train_scaled['TV'], y=y_train )
plt.xlabel('TV')
plt.ylabel('sales')
plt.show()
````

**OBS:** O formato de "cone" apresentado na figura √© um indicativo que a suposi√ß√£o de homocedasticidade (vari√¢ncia constante) pode estar violada.

Se essa suposi√ß√£o √© violada:

- Os estimadores de $\beta_0$ e $\beta_1$ (os coeficientes) ainda ser√£o n√£o-viesados (o valor m√©dio deles √© correto), mas:

  - As infer√™ncias (como p-valores e intervalos de confian√ßa) podem ser enganosas.


Um dos poss√≠veis tratamentos para este problema √© utilizar uma transforma√ß√£o:

- A transforma√ß√£o logar√≠tmica costuma ajudar nesse tipo de situa√ß√£o.

- Uma alternativa mais geral, √© a transforma√ß√£o de Box-Cox.

> Outra possibilidade seria modelar a vari√¢ncia por meio de um segundo processo de regress√£o, a qual √© estudada na disciplina Modelos Lineares Generalizados (MLG).

Na figura abaixo, vamos utilizar a transforma√ß√£o logar√≠tmica.

````python
plt.figure(figsize=(6, 3))
plt.scatter(x=X_train_scaled['TV'], y= np.log(y_train))
plt.xlabel('TV')
plt.ylabel('log_sales')
plt.show()
````

Essa ultima figura sugere que o modelo abaixo √© mais apropriado que o anterior:
$$
\log( sales_i ) = \beta_0 + \beta_1 \times TV_i + …õ_i.
$$

### Ajuste do modelo e res√≠duos

Vamos inicialmente utilizar a biblioteca `sklearn` e na sequ√™ncia vamos repetir o processo via a biblioteca `statsmodels.api`.

> `sklearn`: essa biblioteca √© mais utilizada no contexto de Aprendizado de M√°quina;

> `statsmodels.api`: equivalente ao padr√£o utilizado em R, geralmente preferida na Estat√≠stica.

Afim de verificar a qualidade do modelo, vamos calcular $R^2$ e $R^2_{adj}$, a saber:

|                                         | R¬≤                                                                                       | R¬≤ Ajustado                                                                  |
| --------------------------------------- | ---------------------------------------------------------------------------------------- | ---------------------------------------------------------------------------- |
| **O que √©?**                            | Propor√ß√£o da varia√ß√£o de Y explicada pelo modelo.                                        | Vers√£o "corrigida" do R¬≤ que penaliza o excesso de vari√°veis no modelo.      |
| **F√≥rmula**                             | $R^2 = 1 - \frac{SS_{Res}}{SS_{Tot}}$                                                    | $R^2_{adj} = 1 - \left( \frac{(1 - R^2) \times (n - 1)}{n - p - 1} \right)$  |
| Onde                                    |                                                                                          | n = n√∫mero de observa√ß√µes, p = n√∫mero de vari√°veis preditoras                |
| **Interpreta√ß√£o**                       | Mede o quanto da variabilidade de Y o modelo conseguiu explicar com as vari√°veis usadas. | Ajusta o R¬≤ levando em conta a complexidade do modelo (n√∫mero de vari√°veis). |
| **Cresce se adicionar mais vari√°veis?** | **Sempre aumenta ou se mant√©m igual**, mesmo que a vari√°vel n√£o seja relevante.          | **S√≥ aumenta se a nova vari√°vel realmente melhorar o modelo.**               |
| **Quando usar?**                        | Quando est√° comparando modelos com o **mesmo n√∫mero de vari√°veis**.                      | Quando est√° comparando **modelos com quantidades diferentes de vari√°veis**.  |


#### Via `sklearn`

````python
# Importando o modelo
from sklearn.linear_model import LinearRegression

# Instanciando o modelo
modelo_1 = LinearRegression( fit_intercept=True )

# Ajustando o modelo aos dados de treino
modelo_1.fit( X_train_scaled[['TV']] , np.log(y_train) )
````

````python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Predi√ß√µes ( valores ajustados )
y_pred_log = modelo_1.predict( X_train_scaled[['TV']] )

# R¬≤
r2 = r2_score(np.log(y_train), y_pred_log)

# R¬≤ Ajustado
n = X_train_scaled.shape[0]
p = 1  # N√∫mero de preditores (apenas TV)
r2_adj = 1 - (1 - r2) * (n - 1) / (n - p - 1)

# Erro M√©dio Quadr√°tico
mse = mean_squared_error( np.log(y_train), y_pred_log )
rmse = np.sqrt( mse )

# Print "Resumo b√°sico"
print(f"Intercepto (beta_0): {modelo_1.intercept_}")
print(f"Coeficiente TV (beta_1): {modelo_1.coef_[0]}")
print(f"R¬≤: {r2:.4f}")
print(f"R¬≤ Ajustado: {r2_adj:.4f}")
print(f"RMSE (na escala log): {rmse:.4f}")
````

**Gr√°fico de dispers√£o + linha de regress√£o**

````python
# Gr√°fico de dispers√£o + linha de regress√£o
plt.figure(figsize=(6, 3))
plt.scatter( X_train_scaled['TV'], np.log(y_train), color='blue', label='Dados reais')
plt.plot( X_train_scaled['TV'], y_pred_log, color='red', label='Predi√ß√£o do modelo', alpha=0.7)
plt.xlabel('Investimento em TV (padronizado)')
plt.ylabel('log sales')
plt.title('Regress√£o Linear Simples: Sales vs TV')
plt.legend()
plt.show()
````

**Gr√°fico de res√≠duos vs predi√ß√£o (para verificar homocedasticidade)**

- Queremos ver os res√≠duos espalhados de forma aleat√≥ria ao redor da linha zero.

- Padr√µes ou funil (heterocedasticidade) s√£o sinais de problemas.

````python
residuos_log = np.log(y_train) - y_pred_log

plt.figure(figsize=(6, 3))
plt.scatter(y_pred_log, residuos_log, alpha=0.7)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Valores preditos (log-scale)')
plt.ylabel('Res√≠duos')
plt.title('Res√≠duos vs Preditos (escala log)')
plt.show()
````

#### Via `statsmodels.api`

````python
import statsmodels.api as sm

# Precisa adicionar o intercepto manualmente
X_sm = sm.add_constant(X_train_scaled[['TV']])

modelo_sm = sm.OLS( np.log(y_train), X_sm ).fit()
print(modelo_sm.summary())
````

## Regress√£o Linear Multipla

Neste exemplo, vamos relacionar o logaritmo da vari√°vel resposta (`sales`) com todas as outras vari√°veis, ou seja, vamos utilizar um modelo com a seguinte equa√ß√£o:

$$
\log( sales_i ) = \beta_0 + \beta_1 \times TV_i + \beta_2 \times radio_i + \beta_3 \times newspaper_i + …õ_i,
$$

em que, $(\beta_0, \beta_1,\beta_2,\beta_3) \in \mathbb{R^4}\,$ e $\,…õ_1, \dots, …õ_n$ s√£o i.i.d. Normal$(0,\sigma^2)$.


> OBS: lembre que as covari√°veis $TV, radio, newspaper$ foram padronizadas, ou seja, o modelo proposto vai medir o quanto a varia√ß√£o de um desvio padr√£o em uma dessas vari√°veis afeta o logaritmo de $sales$.

### Ajuste do modelo e res√≠duos

````python
# Importando o modelo
from sklearn.linear_model import LinearRegression

# Instanciando o modelo
modelo_2 = LinearRegression( fit_intercept=True )

# Ajustando o modelo aos dados de treino
modelo_2.fit( X_train_scaled, np.log(y_train) )
````

````python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Predi√ß√µes
y_pred_log = modelo_2.predict(X_train_scaled)

# R¬≤
r2 = r2_score( np.log(y_train), y_pred_log )

# R¬≤ Ajustado
n = X_train_scaled.shape[0]
p = 3  # N√∫mero de preditores
r2_adj = 1 - (1 - r2) * (n - 1) / (n - p - 1)

# Erro M√©dio Quadr√°tico
mse = mean_squared_error( np.log(y_train), y_pred_log )
rmse = np.sqrt(mse)

# Print "Resumo b√°sico"
print(f"Intercepto: {modelo_2.intercept_}")
print(f"Coeficiente TV: {modelo_2.coef_[0]}")
print(f"R¬≤: {r2:.4f}")
print(f"R¬≤ Ajustado: {r2_adj:.4f}")
print(f"RMSE (na escala log): {rmse:.4f}")
````

**Gr√°fico de res√≠duos vs predi√ß√£o (para verificar homocedasticidade)**

- Queremos ver os res√≠duos espalhados de forma aleat√≥ria ao redor da linha zero.

- Padr√µes ou funil (heterocedasticidade) s√£o sinais de problemas.

````python
residuos_log = np.log(y_train) - y_pred_log

plt.figure(figsize=(6, 3))
plt.scatter(y_pred_log, residuos_log, alpha=0.7)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Valores preditos (log-scale)')
plt.ylabel('Res√≠duos')
plt.title('Res√≠duos vs Preditos (escala log)')
plt.show()
````

## Comparando os modelos

**An√°lise *in-sample*:** Comparando o $R^2_{adj}$ calculado para os dois modelos, temos que o modelo de regress√£o multipla teve o maior valor e portanto seria o favorito. No entanto, essa informa√ß√£o se baseia unicamente nos dados de treinamento.

**An√°lise *out-sample*:**  Vamos agora avaliar o desempenho preditivo desses modelos utilizando o conjunto de teste que separamos no in√≠cio do estudo.

  - Essa an√°lise √© especialmente √∫til pois modelos podem estar super-ajustados, o que √© n√£o desej√°vel.

  - Al√©m do mais, tal an√°lise vai permiter ter uma medida mais precisa do desempenho desses modelos fora do conjunto de treino.

> Na c√©lula abaixo, fazemos a predi√ß√£o para o conjunto de teste, inicialmente na escala logaritmica, na sequ√™ncia passamos para a escala real e calculamos o erro de predi√ß√£o. Por fim, calculamos um conjunto de m√©tricas afim de comparar os dois modelos em termos do erro de predi√ß√£o.

````python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Predi√ß√µes para "teste" na escala logaritmica
m1_pred_log = modelo_1.predict( X_test_scaled[['TV']] )
m2_pred_log = modelo_2.predict( X_test_scaled )

# Passando para a escala original
m1_pred = np.exp( m1_pred_log )
m2_pred = np.exp( m2_pred_log )

# Calculando erros de previsao
m1_error = y_test - m1_pred
m2_error = y_test - m2_pred

# M√©tricas para o Modelo 1 (Regress√£o Simples - TV)
m1_mae = mean_absolute_error(y_test, m1_pred)
m1_mse = mean_squared_error(y_test, m1_pred)
m1_rmse = np.sqrt(m1_mse)
m1_r2 = r2_score(y_test, m1_pred)

# M√©tricas para o Modelo 2 (Regress√£o M√∫ltipla - TV + outras vari√°veis)
m2_mae = mean_absolute_error(y_test, m2_pred)
m2_mse = mean_squared_error(y_test, m2_pred)
m2_rmse = np.sqrt(m2_mse)
m2_r2 = r2_score(y_test, m2_pred)

# Exibindo as m√©tricas lado a lado
print("===== Modelo 1: Regress√£o Linear Simples (TV) =====")
print(f"MAE: {m1_mae:.4f}")
print(f"MSE: {m1_mse:.4f}")
print(f"RMSE: {m1_rmse:.4f}")
print(f"R¬≤: {m1_r2:.4f}")

print("\n===== Modelo 2: Regress√£o Linear M√∫ltipla (TV + outras) =====")
print(f"MAE: {m2_mae:.4f}")
print(f"MSE: {m2_mse:.4f}")
print(f"RMSE: {m2_rmse:.4f}")
print(f"R¬≤: {m2_r2:.4f}")
````

>> Logo, o modelo de regress√£o multipla se saiu melhor de acordo com todas as m√©tricas.

No que segue, vamos analisar os erros de predi√ß√£o por meio de gr√°ficos.

**Boxplot dos erros absolutos**

- Neste gr√°fico, o modelo que apresentar erros distribuidos em valores menores √© o melhor.

````python
erro_abs_df = pd.DataFrame({
    'Erro Absoluto': np.concatenate([np.abs(m1_error), np.abs(m2_error)]),
    'Modelo': ['Simples'] * len(m1_error) + ['M√∫ltiplo'] * len(m2_error)
})

plt.figure(figsize=(6,4))
sns.boxplot(x='Modelo', y='Erro Absoluto', data=erro_abs_df)
plt.title('Boxplot dos Erros Absolutos')
plt.show()
````

**Distribui√ß√£o acumulada dos erros absolutos**

- Neste gr√°fico, o modelo que cresce mais rapidamente √© o melhor, pois indica que os erros absolutos est√£o distribuidos em valores mais baixos.

# Exerc√≠cio 2

Para a vari√°vel resposta `mpg`, selecione uma covari√°vel e ajuste um modelo de regress√£o linear simples. Fa√ßa an√°lise de res√≠duos e analise o $R^2_{adj}$.

````python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = sns.load_dataset('mpg')
df = df.dropna()

y = df['mpg']
X = df[['weight']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns, index=X_train.index)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns, index=X_test.index)


modelo_simples = LinearRegression()
modelo_simples.fit(X_train_scaled, y_train)

y_pred_train = modelo_simples.predict(X_train_scaled)
r2 = r2_score(y_train, y_pred_train)

n = X_train_scaled.shape[0] # n√∫mero de observa√ß√µes no treino
p = X_train_scaled.shape[1] # n√∫mero de preditores (neste caso, 1)
r2_adj = 1 - (1 - r2) * (n - 1) / (n - p - 1)

print(f"R¬≤ do modelo simples (treino): {r2:.4f}")
print(f"R¬≤ Ajustado do modelo simples (treino): {r2_adj:.4f}")
print(f"Coeficiente (Weight): {modelo_simples.coef_[0]:.4f}")
print(f"Intercepto: {modelo_simples.intercept_:.4f}")

residuos_train = y_train - y_pred_train

plt.figure(figsize=(8, 4))
plt.scatter(y_pred_train, residuos_train, alpha=0.7)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Valores preditos (treino)')
plt.ylabel('Res√≠duos (treino)')
plt.title('Res√≠duos vs Preditos (Modelo Simples)')
plt.show()

plt.figure(figsize=(6, 4))
sns.histplot(residuos_train, kde=True, bins=20)
plt.title('Distribui√ß√£o dos Res√≠duos (Modelo Simples)')
plt.xlabel('Res√≠duos')
plt.show()
````

# Exerc√≠cio 3

Para a vari√°vel resposta `mpg`, ajuste um modelo de regress√£o multipla considerando as seguintes covari√°veis: `horsepower, weight, displacement, acceleration`. Fa√ßa an√°lise de res√≠duos e analise o $R^2_{adj}$. Compare o $R^2_{adj}$ desse modelo o do Exerc√≠cio 2.


````python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = sns.load_dataset('mpg')
df = df.dropna()

y = df['mpg']
X_multi = df[['horsepower', 'weight', 'displacement', 'acceleration']] # Seleciona m√∫ltiplas colunas

X_train_multi, X_test_multi, y_train_multi, y_test_multi = train_test_split(X_multi, y, test_size=0.3, random_state=42)

scaler_multi = StandardScaler()
X_train_scaled_multi = scaler_multi.fit_transform(X_train_multi)
X_test_scaled_multi = scaler_multi.transform(X_test_multi)

X_train_scaled_multi = pd.DataFrame(X_train_scaled_multi, columns=X_train_multi.columns, index=X_train_multi.index)
X_test_scaled_multi = pd.DataFrame(X_test_scaled_multi, columns=X_test_multi.columns, index=X_test_multi.index)

modelo_multiplo = LinearRegression()
modelo_multiplo.fit(X_train_scaled_multi, y_train_multi)

y_pred_train_multi = modelo_multiplo.predict(X_train_scaled_multi)
r2_multi = r2_score(y_train_multi, y_pred_train_multi)

n_multi = X_train_scaled_multi.shape[0] # n√∫mero de observa√ß√µes no treino
p_multi = X_train_scaled_multi.shape[1] # n√∫mero de preditores (neste caso, 4)
r2_adj_multi = 1 - (1 - r2_multi) * (n_multi - 1) / (n_multi - p_multi - 1)

print("===== M√©tricas do Modelo de Regress√£o Linear M√∫ltipla =====")
print(f"R¬≤ do modelo m√∫ltiplo (treino): {r2_multi:.4f}")
print(f"R¬≤ Ajustado do modelo m√∫ltiplo (treino): {r2_adj_multi:.4f}")
print("Coeficientes:")
for i, col in enumerate(X_train_multi.columns):
    print(f"  {col}: {modelo_multiplo.coef_[i]:.4f}")
print(f"Intercepto: {modelo_multiplo.intercept_:.4f}")

residuos_train_multi = y_train_multi - y_pred_train_multi

plt.figure(figsize=(8, 4))
plt.scatter(y_pred_train_multi, residuos_train_multi, alpha=0.7)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Valores preditos (treino)')
plt.ylabel('Res√≠duos (treino)')
plt.title('Res√≠duos vs Preditos (Modelo M√∫ltiplo)')
plt.show()

plt.figure(figsize=(6, 4))
sns.histplot(residuos_train_multi, kde=True, bins=20)
plt.title('Distribui√ß√£o dos Res√≠duos (Modelo M√∫ltiplo)')
plt.xlabel('Res√≠duos')
plt.show()

try:
    print("\n===== Compara√ß√£o do R¬≤ Ajustado =====")
    print(f"R¬≤ Ajustado do Modelo Simples (Peso): {r2_adj:.4f}")
    print(f"R¬≤ Ajustado do Modelo M√∫ltiplo (Horsepower, Weight, Displacement, Acceleration): {r2_adj_multi:.4f}")

    if r2_adj_multi > r2_adj:
        print("O modelo m√∫ltiplo tem um R¬≤ ajustado maior do que o modelo simples.")
    elif r2_adj_multi < r2_adj:
        print("O modelo simples tem um R¬≤ ajustado maior do que o modelo m√∫ltiplo.")
    else:
        print("Os modelos t√™m o mesmo R¬≤ ajustado.")

except NameError:
    print("\nN√£o foi poss√≠vel comparar o R¬≤ ajustado diretamente. Por favor, execute o c√≥digo do Exerc√≠cio 2 primeiro.")
````
